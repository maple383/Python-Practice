```python
import matplotlib as mpl
mpl.use('Agg')


#from matplotlib import style
#style.use('seaborn')


import matplotlib.pyplot as plt

# 四個月份
labels = ["Jun","Jul","Aug","Sep"]
sizes = [20, 30, 40, 10]

fig,ax = plt.subplots(1,2)

ax[0].bar(labels, sizes, color='b',width=0.5)


# 圓餅圖顏色
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = [0,0,0.1,0]

# 數據顯示格式
autopct = '%.1f%%'

ax[1].pie(sizes, labels=labels, colors=colors, autopct = autopct , explode = explode)

plt.show()
```
