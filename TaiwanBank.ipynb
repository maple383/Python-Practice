```python
'''
台灣銀行歷史匯率查詢，與下載 Excel檔案查看內容
2020-05-20
'''



from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import time

# 開啟瀏覽器
def openChrome():

    # download.prompt_for_download": False (禁止跳出視窗) ， download.default_directory 設定檔案下載位置
    prefs = {"download.prompt_for_download": False, 'download.default_directory': r"C:\Users\user"}

    options = Options()
    options.add_experimental_option('prefs', prefs)

    driver_Chrome = webdriver.Chrome('/Users/user/chromedriver',chrome_options=options)
    
    return driver_Chrome



def Operations(driver):

    url = 'https://rate.bot.com.tw/xrt/history?lang=zh-TW'
    
    driver.get(url)
    driver.set_window_position(1,0)
    driver.maximize_window()  # 視窗最大化

    # 按下 ESC鍵取消訊息視窗
    #webdriver.ActionChains(driver).send_keys(Keys.ESCAPE).perform()

    # 點擊 "最近半年走勢圖"
    driver.find_element_by_xpath("//div[@class='controls']/label[3]").click()

    # 選擇查詢幣別 - (人民幣)
    driver.find_element_by_xpath("//div[@class='controls']/select[@name='currency']/option[@value='CNY']").click()

    # 比較幣別 - (美金)
    driver.find_element_by_xpath("//div[@class='controls']/select[@name='compare']/option[@value='USD']").click()

    # 點選 '現金' or '即期'
    driver.find_element_by_xpath("//div[@class='controls']/label/input[@title='現金']").click()

    # 查詢
    driver.find_element_by_xpath("//div[@class='form-actions']/a").click()
    time.sleep(5)

    # 下載 Excel檔案到指定位置
    driver.find_element_by_xpath("//div[@class='text-center']/div/a[@class='btn']").click()
    
    # 視窗最小化
    driver.minimize_window()
    
def data_read():
    
    import pandas as pd
    import numpy as np

    file_name = input('請輸入要讀取檔案名稱+副檔名')
    df = pd.read_csv(file_name)

    new_clo = []
    for i in (df.columns):
        new_clo.append(i)


    # 重新命名標頭欄
    df = pd.read_csv(file_name,names = new_clo)

    # 去除第一欄
    df = df.drop([0])
    data = df[['資料日期','幣別','匯率','現金','即期','匯率.1','現金.1','即期.1']].head(10)
    
    print(data)
    
    
    

if __name__ == '__main__':
    
    driver = openChrome()
    Operations(driver)
    file = data_read()
    

```
