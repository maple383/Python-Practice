```python
#自由時報即時頭條新聞

from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests

html= requests.get('https://www.ltn.com.tw/')

if html.status_code == requests.codes.ok:
    bs = BeautifulSoup(html.text,'html.parser')
    
    titles = bs.find_all('a',{'class':'title','target':'_self'})  # {'target':'_self'}消除廣告部分

    #print(titles)
    for i in titles :
        
        print(i.get_text().strip())
        print(i['href'],'\n')
```
