```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,6)  # x = 1 ~ 5

#也可以 x +0.4 ， 長條圖往右 +0.4 個數值 (1.4 , 2.4 , 3.4 , 4.4 , 5.4)
plt.bar(x - 0.4 , [3,10,8,12,6], width = 0.4 , ec = 'none', fc = "#33FFFF")   # x - 0.4 表式線條顯示在  x = 0.6的位置( 0.6 , 1.6 , 2.6 , 3.6 , 4.6 )，這樣顯示間距較漂亮 & 不會跟下面的圖線重疊
plt.bar(x , [6,3,12,5,8], width = 0.4 , ec = 'none', fc = "#800000")
```

# 疊加型資料

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,6)

A = np.random.randint(2,15,5)
B = np.random.randint(2,15,5)
C = np.random.randint(2,15,5)

plt.bar(x ,A, fc = "#0000FF",ec = "none")
plt.bar(x ,B, fc = "#EEEE00",ec = "none",bottom = A)     # botoom 底座是 A ，如果沒有設定會無法顯示
plt.bar(x ,C, fc = "#227700",ec = "none",bottom = A+B)
```

# 直方圖

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(500,50,2000)

# 以等差級數決定方盒的數量
bins = np.linspace(np.ceil(min(data)), np.floor(max(data)),40)
plt.hist(data , bins,color = "#AA0000",alpha = 0.4)

#設定 y 軸刻度
plt.ylim([0,160])

# 設定 x 軸名稱
plt.xlabel('X - value',size = 20)

# 設定圖表標題
plt.title('Norma',size = 20)
plt.show()
```

# 散點圖 Scatter

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 隨機 50個 0~10之間的浮點數
x = np.random.rand(50)*10

# 隨機產生 50個範圍 0~100 的整數
y = np.random.randint(0,101,50)

plt.scatter(x,y, s = 80 , marker = '^' , c = np.random.rand(50), alpha = 0.7)

#設定 x軸的名稱
plt.xlabel('X - value',size = 14)

#設定 y軸的名稱
plt.ylabel('Y - value',size = 14)

#設定 x軸的刻度 0~10
plt.xticks(np.arange(0,11)) 

#設定 y軸的刻度 0~100，間隔 20
plt.yticks(np.arange(0,101,20))

#設定圖表標題
plt.title("Title", fontsize = 20)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 隨機 50個 0~10之間的浮點數
x = np.random.rand(50)*10

# 隨機產生 50個範圍 0~100 的整數
y = np.random.randint(0,101,50)

plt.scatter(x,y, s = 80 , marker = '^' , c = np.random.rand(50), alpha = 0.7)

#設定 x軸的名稱
plt.xlabel('X - value',size = 14)

#設定 y軸的名稱
plt.ylabel('Y - value',size = 14)

#設定 x軸的刻度 0~10
xrange = ['0~2','2~4','4~6','6~8','8~10']
plt.xticks(np.arange(0,11,2)+2,xrange) 

#設定 y軸的刻度 0~100，間隔 20
plt.yticks(np.arange(0,101,20))

#設定圖表標題
plt.title("Title", fontsize = 20)
plt.show()
```
